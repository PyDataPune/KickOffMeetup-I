List comprehensions

```
nums = [54,22,85,66,14,89]
```

```
nums_update = []
```

```
for num in nums:
  nums_update.append(num + 1) # Populate a list with for Loop
```

```
print(nums_update)
```

```
nums_new = [num + 2 for num in nums]
print(nums_new)
```

```
res = [re + 1 for re in range(1,10)]
```

```
print(res)
```

List comprehension in the place of for loops

```
pairs_2 = [(x, y) for x in range(0,2) for y in range(6,8)] # Nested loop by LC
```

```
print(pairs_2)
```

```
[num ** 2 for num in range(10) if num%2 == 1] # LC with conditional expression
```

```
[num ** 2 if num%2 == 1 else 0 for num in range(10)] # LC with conditional expression on output
```

```
p_neg = {num: -num for num in range(9)} # Comprehension in dictionaries
```

```
print(p_neg)
```

```
result = (2 * num for num in range(10))
```

```
for num in result: # printing values from generators
  print(num)
```

```
result1 = (num for num in range(5))
```

```
print(list(result1)) # Generators to List
```

```
result2 = (num for num in range(8)) # Example of Lazy Evaluation
```

```
print(next(result2))
```

```
print(next(result2))
```

```
def num_sequence(n):
  """ Generate values from 0 to n"""
  i = 0
  while i < n:
    yield i
    i += 1
```

```
temp = num_sequence(4)
```

```
print(type(temp))
```

```
for item in temp:
  print(item)
```
